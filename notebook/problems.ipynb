## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import random

def simulate_dice_experiment(num_simulations):
    count_greater_than_7_or_even = 0

    for _ in range(num_simulations):
        # Roll two dice
        dice1 = random.randint(1, 6)
        dice2 = random.randint(1, 6)

        # Calculate the sum
        total_score = dice1 + dice2

        # Check if the sum is greater than 7 or even
        if total_score > 7 or total_score % 2 == 0:
            count_greater_than_7_or_even += 1

    # Calculate the estimated probability
    probability = count_greater_than_7_or_even / num_simulations

    return probability

# Run the experiment 1000 times
num_simulations = 1000
result = simulate_dice_experiment(num_simulations)

print(f"The estimated probability is: {result}")
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
import random

def simulate_ball_experiment(num_simulations):
    count_take_3_white_and_2_red = 0
    count_all_same_color = 0

    for _ in range(num_simulations):
        # Draw 5 balls with replacement
        drawn_balls = random.choices(['white', 'red', 'green'], k=5)

        # Check probability 1: Take 3 white and 2 red
        if drawn_balls.count('white') == 3 and drawn_balls.count('red') == 2:
            count_take_3_white_and_2_red += 1

        # Check probability 2: All are the same color
        if drawn_balls.count(drawn_balls[0]) == 5:
            count_all_same_color += 1

    # Calculate the probabilities
    probability_take_3_white_and_2_red = count_take_3_white_and_2_red / num_simulations
    probability_all_same_color = count_all_same_color / num_simulations

    return probability_take_3_white_and_2_red, probability_all_same_color

# Run the experiment 1000 times
num_simulations = 1000
result1, result2 = simulate_ball_experiment(num_simulations)

print(f"Probability of taking 3 white and 2 red: {result1}")
print(f"Probability of all balls being the same color: {result2}")
```
